## TorchVision

Pretrained models: https://pytorch.org/vision/stable/models.html

```python
import torch
import torchvision

vgg16 = torchvision.models.vgg16(pretrained=True)
print("Original VGG16")
print(vgg16)
```

```python
number_of_layers = 17
pretrained_model = torch.nn.Sequential(vgg16.features[:number_of_layers])
print("Extracted layers")
print(pretrained_model)
```
